```python
#Q1 코드의 결과값은?
a = "Life is too short, you need python"

if "wife" in a : print("wife")
elif "python" in a and 'you' not in a : print("python")
elif "shirt" not in a : print("shirt")
elif "need" in a : print("need")
else : print("none")
```

```python
#Q2 while문을 사용해 1부터 1000까지의 자연수 중 3의 배수의 합을 구해 보자.
result = 0
i = 1
while i <= 1000:
    if i%3==0 :
        result += i
    i += 1
print(result)
```

```python
#2-2 리스트로 3의배수 뽑기
result = []
i = 1
while i <= 1000:
    if i%3==0 :
        result.append(i)
    i += 1
print(result)
```

```python
#Q3 while문을 사용하여 다음과 같이 별(*)을 표시하는 프로그램을 작성해 보시오.
"""
[다음]
*
**
***
****
*****
"""
i = 0
while i < 5 :
    i+=1
    print("*"*i)
```

```python
#Q4 for문을 사용해 1부터 100까지의 숫자를 출력해 보자.
for i in range(1,101):
    print(i)
```

```python
"""
Q5 A학급에 총 10명의 학생이 있다. 이 학생들의 중간고사 점수는 다음과 같다.
[70,60,55,75,95,90,80,80,85,100]
for문을 사용하여 A학급의 평균점수를 구해보자.
"""
A = [70,60,55,75,95,90,80,80,85,100]
sum=0
for num in A :
    sum+=num
mean = sum/len(A)
print(mean)
```

```python
"""
리스트 중에서 홀수에만 2를 곱하여 저장하는 다음 코드가 있다.
numbers = [1,2,3,4,5]

result = []
for n in numbers:
    if n % 2==.1:
        result.append(n*2)
위 코드를 리스트 내포를 사용하여 표현해보자.
"""
result = [num*2 for num in [1,2,3,4,5] if num%2==1]
print(result)
```
